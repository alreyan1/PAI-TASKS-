# Introduction and Definition

```python
print("\n=== Knowledge Representation in Artificial Intelligence ===\n")

# Definition
print("1. Definition:")
print("Knowledge Representation (KR) is a way to store and organize information so that a computer can understand and use it.")
print("It helps machines to think, reason, and make decisions.\n")
```

# Types of Knowledge

```python
# Types of Knowledge
print("2. Types of Knowledge:")

print("\n- Declarative Knowledge:")
print("  Facts about the world.")
print("  Example: Water is a liquid.")

print("\n- Procedural Knowledge:")
print("  Steps to perform a task.")
print("  Example: Steps to make tea.")

print("\n- Semantic Knowledge:")
print("  Meaning and relationships.")
print("  Example: A cat is an animal.")

print("\n- Episodic Knowledge:")
print("  Personal experiences.")
print("  Example: I visited Paris in 2020.")

print("\n- Metaknowledge:")
print("  Knowledge about knowledge.")
print("  Example: I know that I don’t know everything.\n")
```

# Methods of Knowledge Representation

```python
# Methods of KR
print("3. Methods of Knowledge Representation:")

print("\n- Logical Representation:")
print("  Uses logic rules.")
print("  Example: If it is raining, the ground is wet.")

print("\n- Semantic Networks:")
print("  Uses graphs (nodes and connections).")
print("  Example: Dog -> Animal")

print("\n- Frames:")
print("  Uses structured data like objects.")
print("  Example: Car with attributes like color, brand, wheels.")

print("\n- Production Rules:")
print("  IF-THEN rules.")
print("  Example: IF fever THEN flu.")

print("\n- Ontologies:")
print("  Defines relationships between concepts.")
print("  Example: Flu has symptom fever.\n")
```

# Prolog Example

```python
# Prolog Example
print("4. Prolog Example:")

print("\nFacts:")
print("human(socrates).")

print("\nRule:")
print("mortal(X) :- human(X).")

print("\nQuery:")
print("?- mortal(socrates).  -> true\n")
```
