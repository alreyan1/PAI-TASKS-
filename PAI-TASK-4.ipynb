```python
def is_safe(board, row, col, n):
    for i in range(row):
        if board[i] == col:
            return False
        if abs(board[i] - col) == abs(i - row):
            return False
    return True

def solve(board, row, n):
    if row == n:
        return True
    for col in range(n):
        if is_safe(board, row, col, n):
            board[row] = col
            if solve(board, row + 1, n):
                return True
            board[row] = -1
    return False

def print_board(board, n):
    print()
    for row in range(n):
        line = ""
        for col in range(n):
            if board[row] == col:
                line += " Q "
            else:
                line += " . "
        print(line)
    print()

n = int(input("Enter the number of queens: "))
board = [-1] * n

if solve(board, 0, n):
    print("Solution found!")
    print("Queen positions (row: col):")
    for i in range(n):
        print(f"  Row {i + 1}: Column {board[i] + 1}")
    print_board(board, n)
else:
    print("No solution exists.")
```
