```python
visited = set()

def dfs(jug1, jug2, jug1_cap, jug2_cap, goal, path):
    if (jug1, jug2) in visited:
        return False
    
    visited.add((jug1, jug2))
    path.append((jug1, jug2))

    if jug1 == goal or jug2 == goal:
        return True

    states = []

    states.append((jug1_cap, jug2, "Fill Jug 1"))
    states.append((jug1, jug2_cap, "Fill Jug 2"))
    states.append((0, jug2, "Empty Jug 1"))
    states.append((jug1, 0, "Empty Jug 2"))

    pour = min(jug1, jug2_cap - jug2)
    states.append((jug1 - pour, jug2 + pour, "Pour Jug 1 to Jug 2"))

    pour = min(jug2, jug1_cap - jug1)
    states.append((jug1 + pour, jug2 - pour, "Pour Jug 2 to Jug 1"))

    for next_jug1, next_jug2, rule in states:
        if (next_jug1, next_jug2) not in visited:
            print(f"State: ({next_jug1}, {next_jug2})  -->  {rule}")
            if dfs(next_jug1, next_jug2, jug1_cap, jug2_cap, goal, path):
                return True

    path.pop()
    return False


path = []
print("Start State: (0, 0)")
dfs(0, 0, 4, 3, 2, path)
print("\nSolution Path:")
for state in path:
    print(state)
```
